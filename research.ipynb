```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
df_train
```

```python
sns.countplot(x='label',data=df_train)
```

```python
df_train['word_count'] = df_train['sms'].apply(lambda x: len(str(x).split()))
print(df_train[df_train['label']==1]['word_count'].mean()) #Spam
print(df_train[df_train['label']==0]['word_count'].mean()) #Not Spam
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(10,4))
train_words=df_train[df_train['label']==1]['word_count']
ax1.hist(train_words,range=(0,75),color='red')
ax1.set_title('Spam Texts')
train_words=df_train[df_train['label']==0]['word_count']
ax2.hist(train_words,range=(0,75),color='green')
ax2.set_title('Non-spam Texts')
fig.suptitle('Words per text message')
plt.show()
```

```python
df_train['char_count'] = df_train['sms'].apply(lambda x: len(str(x)))
print("Spam Texts")
print(df_train[df_train['label']==1]['char_count'].mean()) # Spam
print("")
print("Non-Spam Texts")
print(df_train[df_train['label']==0]['char_count'].mean()) # Non-Spam
```
